```python
import pandas as pd
```

```python
#open the file and display basic info
ted = pd.read_csv('ted.csv')
ted.info()
```

```python
#create a dictionary where each title is assigned a number
titleN={}
index=0
for title in ted['title']:
    titleN[title]=index+1
    index+=1
#add a new column to the data frame. The column contains the number assigned to each title.
ted['title number']=ted['title'].map(titleN)
ted.head()
```

```python
#print the possible values of the variable containing missing values
print(ted['speaker_occupation'].drop_duplicates())
```

```python
#create a dataframe containing oly rows with missing values
ted_null=ted[ted['speaker_occupation'].isnull()]
ted_null.info()
```

```python
#remove rows with null values from the original dataframe
ted=ted.dropna()
ted.info()
```

```python
#find the year for each talk
import datetime
```

```python
pd.to_datetime(ted['film_date'],unit='s')
```

```python
#change 'film_date' entry to only contain years
ted['film_date']=pd.to_datetime(ted['film_date'],unit='s').dt.year
```

```python
ted.head()
```

```python
#extract list of years
years=list(ted['film_date'].drop_duplicates())
years.sort()
```

```python
#find number of talks and views per year
views={}
films={}
for year in years:
    count = ted['views'].loc[ted['film_date']==year].sum()
    views[year]=count
    films[year]=len(ted[ted['film_date']==year])
```

```python
#create table with new information 
ted_in_years=pd.DataFrame({'N_talks':films,'N_views':views})
```

```python
ted_in_years
```

```python
#find ratio between number of views and number of talks
ted_in_years['views_talks_ratio']=ted_in_years['N_views']/ted_in_years['N_talks']
```
